```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d","price_change_percentage_30d","price_change_percentage_60d","price_change_percentage_200d","price_change_percentage_1y"]])
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(market_data_scaled, columns=["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d","price_change_percentage_30d","price_change_percentage_60d","price_change_percentage_200d","price_change_percentage_1y"])

# Copy the crypto names from the original DataFrame
df_market_data_scaled["coin_id"]=df_market_data.index

# Set the coin_id column as index
df_market_data_scaled=df_market_data_scaled.set_index("coin_id")

# Display the scaled DataFrame
df_market_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k=list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia=[]

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
curve1=df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k)
curve1
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4 is the best value for k (i used a few random stats to choose)

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
k_4 = model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
k_4
```

```python
# Create a copy of the scaled DataFrame
market_data_scaled_predictions_df = df_market_data_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
market_data_scaled_predictions_df['predictied']=k_4

# Display the copy of the scaled DataFrame
market_data_scaled_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
graph1=market_data_scaled_predictions_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="predictied",
    hover_cols=["coin_id"],
    ).opts(yformatter="%.0f")
graph1
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca=PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
coin_pca = pca.fit_transform(df_market_data_scaled)

# View the scaled PCA data
coin_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 89.50%

```python
# Create a new DataFrame with the PCA data.
coin_pca_df = pd.DataFrame(
    coin_pca,
    columns=["PCA1", "PCA2","PCA3"]
)

# Copy the crypto names from the original scaled DataFrame
coin_pca_df["coin_id"]=df_market_data.index

# Set the coin_id column as index
coin_pca_df=coin_pca_df.set_index("coin_id")

# Display the scaled PCA DataFrame
coin_pca_df.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
pca_k=list(range(1,11))
```

```python
# Create an empty list to store the inertia values
pca_inertia=[]

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data_scaled)
    pca_inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {"k": pca_k, "inertia": pca_inertia}
# Create a DataFrame with the data to plot the Elbow curve
pca_df_elbow = pd.DataFrame(pca_elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
curve2=pca_df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="PCA Elbow Curve", 
    xticks=k)
curve2
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer seems like 4


* **Question:** Does it differ from the best k value found using the original data? 

  * **Answer:**   It is the same but a sharper drop

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
model.fit(coin_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
k_4 = model.predict(coin_pca_df)

# Print the resulting array of cluster values.
k_4
```

```python
# Create a copy of the scaled PCA DataFrame
coin_pca_predictions_df = coin_pca_df.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
coin_pca_predictions_df["predicted"] = k_4

# Display the copy of the scaled PCA DataFrame
coin_pca_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
graph2=coin_pca_predictions_df.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="predicted",
    hover_cols=["coin_id"]
).opts(yformatter="%.0f")
graph2
```
